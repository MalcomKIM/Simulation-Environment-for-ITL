```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['CoppeliaSim','Gazebo','Webots','Unity']
```

```python
cube_cpu = []
sphere_cpu = []
for pe in labels:
    path = 'Simulators Comparison/CPU Useage/{}_cube.csv'.format(pe)
    data = np.genfromtxt(path, delimiter=',')
    cube_cpu.append(round(np.mean(data[:,1]),2))
    
    path = 'Simulators Comparison/CPU Useage/{}_ball.csv'.format(pe)
    data = np.genfromtxt(path, delimiter=',')
    sphere_cpu.append(round(np.mean(data[:,1]),2))

print(cube_cpu)
print(sphere_cpu)
```

```python
x = np.arange(len(labels))  # the label locations
width = 0.45  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, cube_cpu, width, label='Stacking Cubes')
rects2 = ax.bar(x + width/2, sphere_cpu, width, label='Stacking Spheres')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('CPU Usuage (%)')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
plt.ylim(0, 140)

plt.savefig('Figures/CPU.pdf')
plt.show()
```
