```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
physics_engines = ['Bullet', 'ODE', 'PhysX','Newton']
```

```python
0.5 * -9.81
```

```python
for pe in physics_engines:
    path = 'Simulators Comparison/{}/Integrator.csv'.format(pe)
    data = np.genfromtxt(path, delimiter=',')
    print(pe)
    error = 0
    for row in data:
        t = row[0]
        p = row[1]
        e = 0.5 * -9.81 * t * t
        error += np.abs(p-e)
    print(error)
```

# Restitution

$C_{R} = \sqrt{\frac{h}{H}}$\
$C_{R}$ is the coefficient of restitution\
$h$ is the bounce height \
$H$ is the drop height

```python
for pe in physics_engines:
    path = 'Simulators Comparison/{}/Restitution-0.1.csv'.format(pe)
    data = np.genfromtxt(path, delimiter=',')
    plt.plot(data[:,0],data[:,1],label = pe)

plt.axhline(y=0.01, color='black', ls='--', label='Ideal')
plt.xlim(0, 2.1)
plt.ylim(-0.1, 1.1)
plt.title('Restitution = 0.1')
plt.xlabel('Time')
plt.ylabel('Height')
plt.grid()
plt.legend()
plt.savefig('Figures/Restitution-0.1.pdf')
plt.show()
```

# Friction

$\mu = tan\theta$\
$\mu$ is the friction coefficient\
$\theta$ is the angle of the slope

```python
import itertools
marker = itertools.cycle(('s', '^', 'o', 'D')) 
for pe in physics_engines:
    path = 'Simulators Comparison/{}/Friction.csv'.format(pe)
    data = np.genfromtxt(path, delimiter=',')
    plt.plot(data[:,0],data[:,1],marker=next(marker),label = pe)

plt.title('Friction')
plt.xlabel('μ')
plt.ylabel('tanθ')
plt.xticks(np.arange(0, 1.1, 0.1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid()
plt.legend()
plt.savefig('Figures/Friction.pdf')
plt.show()
```

```python
for pe in physics_engines:
    path = 'Simulators Comparison/{}/Collision.csv'.format(pe)
    data = np.genfromtxt(path, delimiter=',')
    plt.plot(data[:,0],data[:,1],label = pe)

plt.axvline(x=0.15, color='black', ls='--')
plt.axvline(x=1.55, color='black', ls='--')
plt.title('Accumulated penetration error')
plt.xlabel('Time')
plt.ylabel('Error')
plt.grid()
plt.legend()
plt.savefig('Figures/Collision.pdf')
plt.show()
```
